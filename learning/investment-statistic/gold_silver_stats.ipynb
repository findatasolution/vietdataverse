```python
start_date = "2000-01-01"
end_date = pd.to_datetime("today").strftime("%Y-%m-%d")
```

```python
gold = yf.download("GC=F", start=start_date, end=end_date)
silver = yf.download("SI=F", start=start_date, end=end_date)
```

```python
gold.columns = gold.columns.droplevel(1)
gold = gold.reset_index()
gold.columns = gold.columns.str.lower()
gold.date = pd.to_datetime(gold.date)
gold= gold.sort_values(by=['date'])

silver.columns = silver.columns.droplevel(1)
silver = silver.reset_index()
silver.columns = silver.columns.str.lower()
silver.date = pd.to_datetime(silver.date)
silver= silver.sort_values(by=['date'])

combine = gold.merge(silver, on='date', how='outer', suffixes=('_gold', '_silver'))
```
